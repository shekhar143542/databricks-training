## SELECT

```python
#1. Display emp_name as employee_name.
df.select(F.col("emp_name").alias("employee_name")).show()

#2. Display salary as monthly_salary.
df.select(F.col("salary").alias("monthly_salary")).show()

#3. Display department as dept.
df.select(F.col("department").alias("dept")).show()

#4. Display joining_date as doj.
df.select(F.col("joining_date").alias("doj")).show()

#5. Select emp_name as name and city as location.
df.select(F.col("emp_name").alias("name"),F.col("city").alias("location")).show()

#6. Display designation as job_role.
df.select(F.col("designation").alias("job_role")).show()

#7. Display age as employee_age.
df.select(F.col("age").alias("employee_age")).show()

#8. Select multiple columns using aliases.
df.select(F.col("emp_name").alias("name"),F.col("city").alias("location"),F.col("salary").alias("emp_salary"),F.col("department").alias("emp_dept")).show()

#9. Display salary as emp_salary and department as emp_dept.
df.select(F.col("salary").alias("emp_salary"),F.col("department").alias("emp_dept")).show()

#10. Display city as work_location.
df.select(F.col("city").alias("work_location")).show()
```

## WITHCOLUMN
This section contains examples of creating new columns or modifying existing columns.

```python
#1. Create bonus column as 10% of salary.
df.withColumn("bonus",F.col("salary")*0.1).show()
#2. Create annual_salary column.
df.withColumn("annual_salary",F.col("salary")*12).show()
#3. Create tax column as 5% of salary.
df.withColumn("tax",F.col("salary")*0.05).show()
#4. Create updated_salary by adding 5000.
df.withColumn("updated_salary",F.col("salary")+5000).show()
#5. Create salary_category column:High,Medium,Low
df.withColumn("salary_category",F.when(F.col("salary")>80000,"High").when(F.col("salary")>50000,"Medium").otherwise("Low")).show()
#6. Create age_group column:Young,Adult
df.withColumn("age_group",F.when(F.col("age")<30,"Young").otherwise("Adult")).show()
#7. Create location column by combining city and department.
df.withColumn("location",F.concat(F.col("city"),F.lit(","),F.col("department"))).show()
#8. Create increment_salary column with 15% hike.
df.withColumn("increment_salary",F.col("salary")*1.15).show()
#9. Create experience_status column based on joining year.
df.withColumn("experience_status",F.when(F.year(F.col("joining_date"))<2016,"Experienced").otherwise("New")).show()
#10. Create name_length column using employee name.
df.withColumn("name_length",F.length(F.col("emp_name"))).show()
#11. Create is_high_salary column using condition.
df.withColumn("is_high_salary",F.when(F.col("salary")>80000,"High").otherwise("Low")).show()
#12. Create joining_year column from joining_date.
df.withColumn("joining_year",F.year(F.col("joining_date"))).show()
#13. Create salary_after_tax column.
df.withColumn("salary_after_tax",F.col("salary")*0.95).show()
#14. Create department_code column.
df.withColumn("department_code",F.concat(F.lit("dept_"),F.col("department"))).show()
#15. Create double_salary column.
df.withColumn("double_salary",F.col("salary")*2).show()
```

## TYPECASTING
This section shows how to convert columns from one datatype to another.

```python
#1. Convert salary to string datatype.
df.withColumn("salary",F.col("salary").cast("string")).show()
#2. Convert age to double datatype.
df.withColumn("age",F.col("age").cast("double")).show()
#3. Convert joining_date to date datatype.
df.withColumn("joining_date",F.col("joining_date").cast("date")).show()
#4. Convert emp_id to string datatype.
df.withColumn("emp_id",F.col("emp_id").cast("string")).show()
#5. Convert salary to integer datatype.
df.withColumn("salary",F.col("salary").cast("integer")).show()
#6. Convert age to string datatype.
df.withColumn("age",F.col("age").cast("string")).show()
#7. Convert joining_date to timestamp datatype.
df.withColumn("joining_date",F.col("joining_date").cast("timestamp")).show()
#8. Convert salary to float datatype.
df.withColumn("salary",F.col("salary").cast("float")).show()
#9. Convert emp_id to long datatype.
df.withColumn("emp_id",F.col("emp_id").cast("long")).show()
#10. Convert multiple columns into different datatypes.
df.withColumn("salary",F.col("salary").cast("float")).withColumn("age",F.col("age").cast("string")).withColumn("joining_date",F.col("joining_date").cast("timestamp")).show()
```

## SORT / ORDERBY
This section contains examples of sorting records in ascending or descending order.

```python
#1. Sort employees by salary ascending.
df.orderBy("salary").show()
#2. Sort employees by salary descending.
df.orderBy(F.desc("salary")).show()
#3. Sort employees by age descending.
df.orderBy(F.desc("age")).show()
#4. Sort employees by emp_name ascending.
df.orderBy(F.asc("emp_name")).show()
#5. Sort employees by city and salary descending.
df.orderBy(F.desc("city"),F.desc("salary")).show()
#6. Sort employees by joining_date.
df.orderBy(F.col("joining_date")).show()
#7. Sort employees by department.
df.orderBy(F.col("department")).show()
#8. Sort employees by designation descending.
df.orderBy(F.desc("designation")).show()
#9. Sort employees first by city then age.
df.orderBy(F.col("city"),F.col("age")).show()
#10. Sort employees by salary and limit top 10.
df.orderBy(F.desc("salary")).limit(10).show()
#11. Sort employees by emp_id descending.
df.orderBy(F.desc("emp_id")).show()
#12. Sort employees from IT department by salary descending.
df.filter(F.col("department") == "IT").orderBy(F.desc("salary")).show()
#13. Sort employees by joining_date descending.
df.orderBy(F.desc("joining_date")).show()
#14. Sort employees alphabetically by emp_name.
df.orderBy(F.asc("emp_name")).show()
#15. Sort employees by multiple columns.
df.orderBy(F.col("city"),F.col("salary")).show()
```

## LIMIT
This section shows how to display only a limited number of rows from the DataFrame.

```python
#1. Display first 5 records.
df.limit(5).show()
#2. Display top 10 employees.
df.orderBy(F.desc("salary")).limit(10).show()
#3. Display first 3 employees from IT department.
df.filter(F.col("department") == "IT").orderBy(F.asc("emp_id")).limit(3).show()
#4. Display top 5 highest salary employees.
df.orderBy(F.desc("salary")).limit(5).show()
#5. Display lowest 5 salary employees.
df.orderBy(F.asc("salary")).limit(5).show()
#6. Display first 7 rows after sorting by age.
df.orderBy(F.asc("age")).limit(7).show()
#7. Display first 2 employees from Hyderabad.
df.filter(F.col("city") == "Hyderabad").limit(2).show()
#8. Display first 15 records from dataframe.
df.limit(15).show()
#9. Display top 5 youngest employees.
df.orderBy(F.asc("age")).limit(5).show()
#10. Display first 8 employees after filtering salary > 60000
df.filter(F.col("salary") > 60000).limit(8).show()
```
